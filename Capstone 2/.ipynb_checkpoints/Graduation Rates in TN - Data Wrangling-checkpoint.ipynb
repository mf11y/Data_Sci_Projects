```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./data/18_19_Grad_rate_district.csv')
```

```python
demo = pd.read_excel('./data/18_19_Demo.xlsx')
```

```python
data
```

```python
data = data[(data.grad_count.str.isnumeric())].copy()
```

```python
data.sample(20)
```

```python
data.info()
```

```python
data.grad_count = data.grad_count.astype(int)
```

```python
data.grad_rate = data.grad_rate.astype(float)
```

```python
data.groupby('subgroup').grad_rate.mean().sort_values()
```

```python
data.subgroup.value_counts()
```

```python
data = data[['system','school','subgroup', 'grad_rate', 'grad_count', 'grad_cohort']]
```

```python
data['school_no'] = data.groupby(['system','school']).ngroup()
```

```python
data
```

```python
data.subgroup.unique()
```

```python
#data.drop(columns=['system', 'school'], inplace=True)dd
```

```python
data = data.pivot(index='school_no',columns='subgroup', values='grad_rate').sort_values(by='Economically Disadvantaged', ascending=True)[[
      'White',
      'Hispanic',
      'Black or African American',
      'Male',
      'Female',
      'Economically Disadvantaged',
      'All Students']]
#data[~data.isna().any(axis=1)]
```
