```python
data = data[(data.grad_count.str.isnumeric())].copy()
```

```python
data.sample(20)
```

```python
data.info()
```

```python
data.grad_count = data.grad_count.astype(int)
```

```python
data.grad_rate = data.grad_rate.astype(float)
```

```python
data.groupby('subgroup').grad_rate.mean().sort_values()
```

```python
data.subgroup.value_counts()
```

```python
data = data[['system','school','subgroup', 'grad_rate']]
```

```python
data['school_no'] = data.groupby(['system','school']).ngroup()
```

```python
data = data.pivot(index='school_no',columns='subgroup', values='grad_rate').sort_values(by='Economically Disadvantaged', ascending=True)[[
      'White',
      'Hispanic',
      'Black or African American',
      'Male',
      'Female',
      'Economically Disadvantaged',
      'All Students']]
data[~data.isna().any(axis=1)]
```
