```python
data = data.loc[~data.isna().any(axis=1)].copy()
```

```python
data.Year = data.Year.astype('int')
```

```python
data
```

```python
data = data.loc[(data.Global_Sales > .1) & (data.Year < 2020)].copy()
```

```python
data.Platform.unique()
```

```python
handhelds = ['DS','GBA','PSP', '3DS', 'GB', 'WS','PSV']
data.loc[data.Platform.isin(handhelds), 'System'] = 'Handheld'
data.loc[~data.Platform.isin(handhelds), 'System'] = 'Console'
data.loc[data.Platform == 'PC', 'System'] = 'PC'
```

```python
data
```

```python
data.Platform.value_counts()
```

```python
remove_platforms = ['NG', 'WS', 'SCD', 'TG16']

data = data.loc[~data.Platform.isin(remove_platforms)].copy()
```

```python
# Adds 3 more columns: Hardware generation of platform that game was released on, 
# Type of system game was released on: Handheld or console,
# and platform manufacturer of the platform the game was released on

handheld_gens = {'GB': 4, 'DS':7, 'GBA':6, '3DS':8, 'PSP':7,'PSV':8}

console_gens = {'Wii':7, 'NES':3, 'X360':7, 'PS3':7,'PS2':6, 'SNES':4, \
                'PS4': 8, 'N64': 5, 'PS':5, 'XB':6, 'PC': 0, '2600':2, \
                'XOne': 8, 'GC':6, 'WiiU': 7, 'GEN':4, 'DC':6, 'SAT':5}

platform_creator = {'GB': 'Nintendo', 'DS':'Nintendo', 'GBA':'Nintendo', '3DS':'Nintendo', 'PSP':'Sony','PSV':'Sony', \
                   'Wii':'Nintendo', 'NES':'Nintendo', 'X360':'Microsoft', 'PS3':'Sony','PS2':'Sony', 'SNES':'Nintendo', \
                   'PS4': 'Sony', 'N64': 'Nintendo', 'PS':'Nintendo', 'XB':'Microsoft', 'PC': 'PC', '2600':'Atari', \
                   'XOne': 'Microsoft', 'GC':'Nintendo', 'WiiU': 'Nintendo', 'GEN':'Sega', 'DC':'Sega', 'SAT':'Sega'}

for x in handheld_gens:
    data.loc[data.Platform == x, 'Generation'] = handheld_gens[x]
    
for x in console_gens:
    data.loc[data.Platform == x, 'Generation'] = console_gens[x]
    
for x in platform_creator:
    data.loc[data.Platform == x, 'platform_creator'] = platform_creator[x]
    
data.loc[(data.Platform == 'PC') & (data.Year >= 1990), 'Generation'] = 1
data.loc[(data.Platform == 'PC') & (data.Year >= 2000), 'Generation'] = 2
data.loc[(data.Platform == 'PC') & (data.Year >= 2005), 'Generation'] = 3
data.loc[(data.Platform == 'PC') & (data.Year >= 2010), 'Generation'] = 4
    
data.Generation = data.Generation.astype('int')
```

```python
data.to_csv('./Data/augmented_vg_sales.csv', index=False)
```
